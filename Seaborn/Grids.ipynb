```python
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
iris = sns.load_dataset('iris')
iris.head()
```

```python
iris['species'].unique()
```

```python
sns.pairplot(iris)
```

```python
sns.PairGrid(iris)
```

```python
g = sns.PairGrid(iris)
g.map(plt.scatter)
```

```python
g = sns.PairGrid(iris)
g.map_diag(sns.histplot)
g.map_upper(plt.scatter)
g.map_lower(sns.kdeplot)
```
