```python
import seaborn as sns
%matplotlib inline
```

```python
flights = sns.load_dataset('flights')
```

```python
tips = sns.load_dataset('tips')
```

```python
tips.head()
```

```python
flights.head()
```

## Heatmap

```python
tc =tips.corr(numeric_only= True)
```

```python
tc
```

```python
sns.heatmap(tc)
```

```python
sns.heatmap(tc, annot=True)
```

```python
sns.heatmap(tc, annot=True,cmap='coolwarm')
```

```python
flights
```

```python
fp = flights.pivot_table(index='month', columns='year', values='passengers')
```

```python
fp
```

```python
sns.heatmap(fp)
```

```python
sns.heatmap(fp, cmap='magma')
```

```python
sns.heatmap(fp, cmap='magma', linecolor='white', linewidths=3)
```

```python
sns.heatmap(fp, cmap='coolwarm', linecolor='white', linewidths=1)
```

## Clustermap

```python
sns.clustermap(fp)
```
